```python
data1=data.drop(['YEAR','JAN','FEB','MAR','JUN','JUL','AUG','SEP','OCT','NOV','DEC','ANNUAL'],axis=1)
```

```python
data1['ANNUAL_RAIN'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
```

```python
data1
```

```python
data1['JAN'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['FEB'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['MAR'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['APR'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['MAY'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['JUNE'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['JULY'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['AUG'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['SEP'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['OCT'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['NOV'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
data1['DEC'] = data.apply([lambda row: row.JAN+row.FEB],axis=1)
```

```python
data1
```

```python
cor=data1.corr()
cor
```
